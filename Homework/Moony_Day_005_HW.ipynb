```python
import seaborn as sns
sns.__version__
```

```python
#看NAME_CONTRACT_TYPE跟TARGET的關係
df=app_train.sample(20000)
sns.set(style="whitegrid")
g = sns.catplot(x="NAME_CONTRACT_TYPE", hue="TARGET", data=df,
                height=6, kind="count", palette="muted")
g.despine(left=True)
g.set_ylabels("Count of NAME_CONTRACT_TYPE")
```

```python
#看CODE_GENDER跟TARGET的關係
df=app_train.sample(20000)
sns.set(style="whitegrid")
g = sns.catplot(x="CODE_GENDER", hue="TARGET", data=df,
                height=6, kind="count", palette="muted")
g.despine(left=True)
g.set_ylabels("Count of CODE_GENDER")
```

```python
#看OCCUPATION_TYPE跟TARGET的關係
df=app_train.sample(20000)
sns.set(style="whitegrid",font_scale=2)
g = sns.catplot(x="OCCUPATION_TYPE", hue="TARGET", data=df,
                height=25, kind="count", palette="muted")
g.despine(left=True)
g.set_ylabels("Count of OCCUPATION_TYPE")
# Rotate x-labels
plt.xticks(rotation=45)
```

```python
#看AMT_INCOME_TOTAL跟TARGET的關係
sns.set(style="whitegrid")
df=app_train.sample(50000)
g = sns.FacetGrid(df, col="TARGET", margin_titles=True,height=7)
bins = np.linspace(0, 700000)
g.map(plt.hist, "AMT_INCOME_TOTAL", color="steelblue", bins=bins)
```
